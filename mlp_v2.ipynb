```python
import pandas as pd 
import numpy as np

import random
```

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
```

```python
%%time
dataset = pd.read_csv(r'label_final_data.csv')

```

```python
dataset
```

```python
df= pd.concat([dataset[dataset['label'] >= 0.5][:100], dataset[dataset['label'] < 0.5][:200]])
df =df.sample(frac=1, random_state=42).reset_index(drop=True) 
df
```

### Split data

```python
X, y = df.drop(columns='label'), df[['label']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
def logfunc(x):
    return 1.0/(1 + np.exp(-x))
```

```python
epoch_count = 100

epsilon = 0.01

I_dim = 41
H_dim = 3

weights_ItoH = np.random.uniform(-1, 1, (I_dim, H_dim))
weights_HtoO = np.random.uniform(-1, 1, H_dim)

preActivation_H = np.zeros(H_dim)
postActivation_H = np.zeros(H_dim)
```

```python
# learning


def learn_perceptron():
    global weights_ItoH, weights_HtoO, preActivation_H, postActivation_H
    #global weights_HtoO
    
    for epoch in range(epoch_count):
        
        err = 0
        for sample in range(len(X_train.iloc[:, 0])):
            if ((sample+1) %500) == 0:
                print('Current sample:',  sample+1)  
            for node in range(H_dim):
                preActivation_H[node] = np.dot(X_train.iloc[sample], weights_ItoH[:, node])
                postActivation_H[node] = logfunc(preActivation_H[node])

            preActivation_O = np.dot(postActivation_H, weights_HtoO)
            postActivation_O = logfunc(preActivation_O)

            Y_real = y_train.iloc[sample][0]

            dn = abs((Y_real - postActivation_O)/Y_real)
            
            #print(dn)

            if dn < epsilon:
                pass

            else:
                q3 = postActivation_O*(1 - postActivation_O)* (Y_real - postActivation_O)
                for node in range(H_dim):
                    q2 = postActivation_H[node]* (1 - postActivation_H[node])* (q3 * weights_HtoO[node])
                    weights_ItoH[:, node] = weights_ItoH[:, node] + q2 * X_train.iloc[sample]

                weights_HtoO = postActivation_H * q3 + weights_HtoO
                
            err += dn
            
        net_error = err/(len(X_train.iloc[:, 0])+1)
        
        if epoch in [0, 49, 99]  :
            print('-------{} epocha-------'.format(epoch+1))
            print('Net error:', net_error, '\n')

    #return weights_ItoH, weights_HtoO
            
```

```python
%%time
learn_perceptron()
```

```python
#recognizing

y_test_arr = y_test.to_numpy()
recog_arr = np.zeros(len(y_test))

def recognizing():
    global weights_ItoH, weights_HtoO, preActivation_H, postActivation_H, \
        y_test_arr, recog_arr
    
    for sample in range(len(X_test.iloc[:, 0])):
        if ((sample+1) %500) == 0:
                print('Current sample:',  sample+1)  
        for node in range(H_dim):
                preActivation_H[node] = np.dot(X_test.iloc[sample], weights_ItoH[:, node])
                postActivation_H[node] = logfunc(preActivation_H[node])

        preActivation_O = np.dot(postActivation_H, weights_HtoO)
        postActivation_O = logfunc(preActivation_O)

        recog_arr[sample] = postActivation_O
    
```

```python
%%time
recognizing()
```
